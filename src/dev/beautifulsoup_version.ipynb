```python
import requests
from bs4 import BeautifulSoup
from transformers import pipeline
```

```python
def summarizeData(data):
    """
    LLM summarize data function
    
    """
    summarizer = pipeline('summarization')
    combinedText= " ".join(data) # combinedata into a  single string
    summary = summarizer(combinedText, max_length=130, min_length=30, do_sample=False)
    return summary[0]['summary_text']
```

```python
def scrapeWebsites(url):
    """
    Scrape websites and get the data
    
    """
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    data = [p.get_text() for p in soup.find_all('p')]
    return soup,  data
```

```python
def scrapeByTag(url, tag):
    """
    Scrape websites by tag
    
    """
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    data = [p.get_text() for p in soup.find_all(tag)]
    return soup,  data
```

```python
def tagFilter(data):
    """
    Filter data by tag
    
    """
    filteredData = []
    for d in data:
        if len(d) > 100:
            filteredData.append(d)
    return filteredData
```
